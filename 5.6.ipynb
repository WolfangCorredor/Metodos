```python
import math
import matplotlib.pyplot as plt
from tabulate import tabulate 
%matplotlib inline

raizReal=1.41906
result=0.7
xl, xu, xr=0.5, 2.0, 0.0
xl2, xu2, xr=0.5,2.0,0.0
cont=-2
evaluar, valores, tabla1, aux1, aux2, tabla2=[],[], [], [],[],[]
fA=1.0
fB=1.0
f1=1.0
f2=1.0
f3=1.0

while(cont<=2):
    x=math.pow(cont, 2)
    now=math.log(x)
    evaluar.append(now)
    valores.append(cont)
    tabla1.append([cont,now])
    cont+=0.02

def metodo1(value):
    arg=math.pow(value, 2)
    ln=math.log(arg)
    return ln
a=1   
while (a<4):
    xr=(xl+xu)/2
    aux1.append(xr)
    fA=metodo1(xl)
    fB=metodo1(xr)
    if ((fA*fB)<0):
        xu=xr
    if((fA*fB)>0):
         xl=xr
    a+=1
    
b=1  
while (b<4):
    f1=metodo1(xl2)
    f2=metodo1(xu2)
    xr2=xu2-(f2*(xl2-xu2))/(f1-f2)
    aux2.append(xr2)
    f3=metodo1(xr2)
    if ((f1*f3)<0):
        xu2=xr2
    if((f1*f3)>0):
         xl2=xr2 
    b+=1 

for i in range(0,3):
    tabla2.append([i+1,aux1[i], aux2[i]])

plt.title("Logaritmo")
plt.axis([0, 2, -2, 4])
plt.ylabel('ln(x^2)')
plt.xlabel('x')
plt.axhline(0, color="black")
plt.plot(valores, evaluar, 'm-', linewidth=2)
plt.plot(raizReal,0.7, 'co', label='X exacta=1.41907',linewidth=2)
plt.legend(loc = 'upper right', numpoints = 1)
plt.show()

print tabulate(tabla2, headers=["Iter","Biseccion","Falsa Posicion"], tablefmt="fancy_grid")
```
