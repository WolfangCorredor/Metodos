```python
matriz,matx,aux,value,i,j,val=[],[],[],0.0,0,0,str("")

def ingreso():
    global var
    global ecu
    var = int(input("Numero de variales:"))
    ecu = int(input("Numero de ecuaciones:"))
    for x in range(0,ecu):
        #print "Columa "
        matriz.append([])
        aux.append([])
        print "ECUACION ",x+1
        for y in range(0,var+1): 
            if y==var:
                val = float(input("B ="))
                matriz[x].append(val)
                aux[x].append(val)
            else:
                pos=y+1
                val = float(input("X%s ="  %(pos)))
                matriz[x].append(val)
                aux[x].append(val)
        #for para mostrar uno por uno
        for num in range(0,var+1):
            pos=num+1
            if num==0:
                cad=str(matriz[x][num])+str('X')+str(pos)
                val=str(val)+str(cad)
            else:
                if num==var:
                    cad=str("=")+str(matriz[x][num])
                    val=str(val)+str(cad)
                else:                    
                    cad=str("+")+str(matriz[x][num])+str('X')+str(pos)
                    val=str(val)+str(cad)
            
        print val

def haciaAdelante():
    for z in range(0,ecu):
        for x in range(z,ecu):
            if x==z:
                value = matriz[z][x]
                for h in range(0,ecu):
                    matriz[z][h] = matriz[z][h]/value
                matriz[z][ecu] = matriz[z][ecu]/value
            else:
                value = matriz[x][z]
                for h in range(0,var):
                    matriz[x][h] = matriz[x][h]-(value*matriz[z][h])
                matriz[x][var] = matriz[x][var]-(value*matriz[z][var])
                
def haciaAtras():
    value = 0.0
    for m in range (var):
        matx.append(0*ecu)
    i=var-1
    j=var-1
    while(i>=0):
        value=matriz[i][var]
        while(j>=i):
            value=value-matx[j]*matriz[i][j]
            j-=1
        matx[i]=value
        i-=1
        j=var-1
    print "\n"
    for i in range (0,var):
        print "El valor de la incognita X%s"%(i+1)," es ",matx[i]
    
def compr():
    print "\n"
    for i in range(0,var):
        total = 0
        for j in range(0,ecu):
            total = total+aux[i][j]*matx[j]
        print "El resultado para la ecuacion",(i+1),"es ",total
        
        
ingreso()
if var==ecu:
    haciaAdelante()
    haciaAtras()
    compr()
else:
    print "No se puede utlizar Gauss Simple porque no es una matriz cuadrada"

```
