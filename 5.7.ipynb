```python
import math
import matplotlib.pyplot as plt 
from tabulate import tabulate
%matplotlib inline

x=-4
xl=1.0
xu=3.0
xr=0.0
valores, funcion, tabla, raices, tabla, Eaprox, tabla1=[],[],[],[],[],[],[]
i=0
true=2.69556

while(x<=4):
    if(x==0):
        continue
    else:
        fx=(0.8-0.3*x)/x
        valores.append(x)
        funcion.append(fx)
        x+=0.1
    tabla.append([valores[i], funcion[i]])
    i+=1

def metodo1(dato):
    fn=(0.8-0.3*dato)/dato
    return fn

b=0
while (b<3):
    f1=metodo1(xl)
    f2=metodo1(xu)
    xr=xu-(f2*(xl-xu))/(f1-f2)
    raices.append(xr)
    f3=metodo1(xr)
    erpv=math.fabs(((true-xr)/true)*100)
    if(b==0):
        Eaprox.append("-----")
        eaprx=100
    else:
        k=b-1
        eaprx=math.fabs(((raices[b]-raices[k])/raices[b])*100)
        Eaprox.append(eaprx)
    tabla1.append([b+1,xr, str(erpv)+" %", str(Eaprox[b])+" %"])
    if ((f1*f3)<0):
        xu=xr
    if((f1*f3)>0):
         xl=xr  
    b+=1
    
plt.title("f(x)")
plt.axis([-4, 5, -6, 6])
plt.ylabel('f(x)')
plt.xlabel('x')
plt.axhline(0, color="black")
plt.plot(valores, funcion, 'r-', linewidth=2)
plt.plot(true,0, 'co', label='Raiz Exacta=2.69556')
plt.legend(loc = 'upper right', numpoints = 1)
plt.show()

print "\nDe manera analitica se obtiene:\n"
print tabulate(tabla, headers=["valor","funcion"], tablefmt="pipe")
print "\nUsando falsa posición con 3 iteraciones\n"
print tabulate(tabla1, headers=["Iteracion","Raiz","Erpv","Eaprox"], tablefmt="pipe")
```
