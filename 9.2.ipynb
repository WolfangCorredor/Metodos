```python
#a)¿Cuáles son las dimensiones de las matrices? Rta= [A]:3x2 , [B]: 3X3, [C]: 3X1, [D]: 2X4, [E]: 3X3, [F]: 2X3, [G]: 1X3
#b)Identifique las matrices cuadrada, columna y renglón. Rta= Cuadrada: [B] y [E]; Columna: [C]; Renglón: [G]
#c)¿Cuáles son los valores de a12,b23,d32,e22,f12 y g12? Rta= a12=7; b23=7; d32=no existe el elemento; e22=2; f12=0; g12=6.

from tabulate import tabulate

A=[[4,7],[1,2],[5,6]]
B=[[4,3,7],[1,2,7],[1,0,4]]
C=[[3],[6],[1]]
D=[[9,4,3,-6],[2,-1,7,5]]
E=[[1,5,8],[7,2,3],[4,0,6]]
F=[[3,0,1],[1,7,3]]
G=[[7,6,4]]
I=[[1,0,0],[0,1,0],[0,0,1]]

#d)Ejecute las operaciones siguientes:

def ingreso():
    print "Ingrese la matriz A (3x2)"
    del A[:]
    for x in range(0,2):
        A.append([])
        for y in range(0,3): 
            val = float(input("valor pos %s,%s "  %(x,y)))
            A[x].append(val)
    print "Ingrese la matriz B (3x3)"
    del B[:]
    for x in range(0,3):
        B.append([])
        for y in range(0,3): 
            val = float(input("valor pos %s,%s "  %(x,y)))
            B[x].append(val) 
    print "Ingrese la matriz C (3x1)"
    del C[:]
    for x in range(0,1):
        C.append([])
        for y in range(0,3): 
            val = float(input("valor pos %s,%s "  %(x,y)))
            C[x].append(val) 
    print "Ingrese la matriz D (2x4)"
    del D[:]
    for x in range(0,4):
        D.append([])
        for y in range(0,2): 
            val = float(input("valor pos %s,%s "  %(x,y)))
            D[x].append(val) 
    print "Ingrese la matriz E (3x3)"
    del E[:]
    for x in range(0,3):
        E.append([])
        for y in range(0,3): 
            val = float(input("valor pos %s,%s "  %(x,y)))
            E[x].append(val) 
    print "Ingrese la matriz F (2x3)"
    del F[:]
    for x in range(0,3):
        F.append([])
        for y in range(0,2): 
            val = float(input("valor pos %s,%s "  %(x,y)))
            F[x].append(val) 
    print "Ingrese la matriz G (1x3)"
    del G[:]
    for x in range(0,3):
        G.append([])
        for y in range(0,3): 
            val = float(input("valor pos %s,%s "  %(x,y)))
            G[x].append(val)

    
ing = int(input("Ingresa\n1.Valores predefinidos para las matrices\n2.Ingresa los valores tu mismo para las matrices"))
if ing==2:
    ingreso()


def condicion(matriza):
    filas=len(matriza)
    columnas=len(matriza[0])
    return filas,"x",columnas
    
print "\nLa matriz A es : ",condicion(A)
print "\nLa matriz B es : ",condicion(B)
print "\nLa matriz C es : ",condicion(C)
print "\nLa matriz D es : ",condicion(D)
print "\nLa matriz E es : ",condicion(E)
print "\nLa matriz F es : ",condicion(F)
print "\nLa matriz G es : ",condicion(G)

def suma(matriz1,matriz2):
    sumaT=[]
    for i in range(len(matriz1)):#filas
        sumaT.append([])
        for j in range(len(matriz1[0])):#columnas
            sumaT[i].append(matriz1[i][j]+matriz2[i][j])
    return sumaT

def resta(matriz1,matriz2):
    restaT=[]
    for i in range(len(matriz1)):#filas
        restaT.append([])
        for j in range(len(matriz1[0])):#columnas
            restaT[i].append(matriz1[i][j]-matriz2[i][j])
    return restaT

def escalar(escalar,matriz):
    product=[]
    for i in range(len(matriz)):#filas
        product.append([])
        for j in range(len(matriz[0])):#columnas
            product[i].append(escalar*matriz[i][j])
    return product

def multiplicacion(a,b):
    product=[]
    for i in range(len(a)):
        product.append([0]*(len(b[0])))
    for i in range(len(a)):
        for j in range(len(b[0])):
            for k in range(len(a[0])):
                product[i][j]+= a[i][k]*b[k][j]
    return product

def transpuesta(a):
    trans=[]
    for i in range(len(a[0])):
        trans.append(len(a)*[0])
        for j in range(len(trans[0])):
            for k in range(len(trans)):
                trans[k][j]=a[j][k]
    return trans

#1)[E]+[B]
print "\n1) La suma de las matrices E+B es: "
print tabulate(suma(E,B), tablefmt="fancy_grid")
#2)[A]+[F]
print "\n2) La suma de las matrices A+F no se puede realizar porque no tienen la misma dimension mxn"
#3)[B]-[E]
print "\n3) La resta de las matrices B-E es: "
print tabulate(resta(B,E), tablefmt="fancy_grid")
print "\n4) Al multiplicar [B] por el escalar 7, tenemos: "
print tabulate(escalar(7,B), tablefmt="fancy_grid")
#5)[E]x[B]
print "\n5) El producto entre las matrices ExB es: "
print tabulate(multiplicacion(E,B), tablefmt="fancy_grid")
#6){C}^T
print "\n6) La transpuesta de {C} es: "
print tabulate(transpuesta(C), tablefmt="fancy_grid")
#7) [B]x[A]
print "\n7)El producto entre las matrices BxA es: "
print tabulate(multiplicacion(B,A), tablefmt="fancy_grid")
#8)[D]^T
print "\n8)La transpuesta de [D] es: "        
print tabulate(transpuesta(D), tablefmt="fancy_grid")
#9)[A]x{C}
print "\n9) El producto [A]x{C} no se puede realizar porque el numero de columnas de A, \nNO coincide con el numero de filas de C"
#10)[I]x[B]
print "\n10) El producto de [I]x[B] es: "            
print tabulate(multiplicacion(I,B), tablefmt="fancy_grid")
#11)[E]^T x [E] --> transpuesta de la matriz E por la matriz E           
print "\n11) El producto de [E]^T x [E] es: "
print tabulate(multiplicacion(transpuesta(E),E), tablefmt="fancy_grid")
#12) {C}^T x {C}--> producto de la transpuesta de la matriz C por la matriz C
print "\n12) El producto de {C}^T x {C} es: "
print tabulate(multiplicacion(transpuesta(C),C), tablefmt="fancy_grid")
```
