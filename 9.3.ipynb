```python

from tabulate import tabulate
matriza = []
matrizb = [] 
matriz1 = [] 
matriz2 = [] 
matriz3 = []
res = []
def ingreso():
    print "MATRIZ A (3X2) "
    for x in range(0,3):
        ec = x+1
        print ("Ecuacion %s" %(ec))
        matriz1.append([])
        for y in range(0,2):
            pos=y+1
            val = float(input("X%s ="  %(pos)))
            matriz1[x].append(val)
    print tabulate(matriz1, tablefmt="fancy_grid")
    print "\n MATRIZ B (2X2) "
    for x in range(0,2):
        ec = x+1
        print ("Ecuacion %s" %(ec))
        matriz2.append([])
        for y in range(0,2):
            pos=y+1
            val = float(input("X%s ="  %(pos)))
            matriz2[x].append(val)
    print tabulate(matriz2, tablefmt="fancy_grid")
    print "\n MATRIZ C (2X2) "
    for x in range(0,2):
        ec = x+1
        print ("Ecuacion %s" %(ec))
        matriz3.append([])
        for y in range(0,2):
            pos=y+1
            val = float(input("X%s ="  %(pos)))
            matriz3[x].append(val)
    print tabulate(matriz3, tablefmt="fancy_grid")
                

def condicion(matriza,matrizb):
    filas=0
    columnas=0
    filas=len(matrizb)
    columnas=len(matriza[0])
          
    #print filas,columnas
    if filas==columnas:
        return True
    else:
        return False
    
def multiplicacion(matriza,matrizb):
    product=[]
    for i in range(len(matriza)):
        product.append([0]*(len(matrizb[0])))
    for i in range(len(matriza)):
        for j in range(len(matrizb[0])):
            for k in range(len(matriza[0])):
                product[i][j]+= matriza[i][k]*matrizb[k][j]
    print tabulate(product, tablefmt="fancy_grid")

def a():
    print "\n a) Multiplicaciones posibles"
    if(condicion(matriz1,matriz2) == True):
        print ("\n [A]x[B]")
        multiplicacion(matriz1,matriz2)
    if(condicion(matriz1,matriz3) == True):
        print ("\n [A]x[C]")
        multiplicacion(matriz1,matriz3)
    if(condicion(matriz2,matriz1) == True):
        print ("\n [B]x[A]")
        multiplicacion(matriz2,matriz1)       
    if(condicion(matriz2,matriz3) == True):
        print ("\n [B]x[C]")
        multiplicacion(matriz2,matriz3) 
    if(condicion(matriz3,matriz1) == True):
        print ("\n [C]x[A]")
        multiplicacion(matriz3,matriz1)
    if(condicion(matriz3,matriz2) == True):
        print ("\n [C]x[B]")
        multiplicacion(matriz3,matriz2) 
        
ingreso()
a()
print ("B: No se pueden multiplicar por que el num de columnas de la pimera matriz no es igual al numero de renglones en la segunda matriz")
print ("C: En productos de matrices no existe propiedad conmutativa, no es lo mismo AXB que BXA")
```
