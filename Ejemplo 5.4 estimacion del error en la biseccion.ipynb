## 2.Compare el desempeño del error relativo porcentual verdadero de los ejemplos 5.4 y 5.5 usando los metodos biseccion y falsa posicion.

Comparando el desempeño de los errores porcentuales verdaderos de los ejemplos 5.4 (clase6) y 5.5 (primer punto de este taller) del chapra, resueltos anteriormente mediante el método de biseccion y falsa posición respectivamente, pude observar que el error decrece mucho más rápido en el método de la falsa posición que en el de la bisección, debido a un esquema más eficiente en el método de la falsa posición para la localización de raíces. Además el método de biseccion aún teniendo más iteracciones , se encontró con un error porcentual relativo verdadero más alto que el de falsa posición.
